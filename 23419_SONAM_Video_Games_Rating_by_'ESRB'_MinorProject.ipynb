Q5.How well can the ESRB rating be predicted based on the presence of different content features? Can you build a classification model to predict the ESRB rating using the available features and evaluate its performance using appropriate metrics?

```python
import pandas as pd
from itertools import combinations
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
file_path = '/content/gdrive/My Drive/DATA/Video_games_esrb_rating.csv'
testfile_path='/content/gdrive/My Drive/DATA/test_esrb.csv'
# Read the CSV file into a Pandas DataFrame
train_pdf = pd.read_csv(file_path)
test_pdf = pd.read_csv(testfile_path)

# split the data
x_train = np.array(train_pdf[train_pdf.columns[2:-1]])
y_train = np.array(train_pdf[train_pdf.columns[-1]])

x_test = np.array(test_pdf[test_pdf.columns[2:-1]])
y_test = np.array(test_pdf[test_pdf.columns[-1]])


#MODELLING
#metric
from sklearn.metrics import accuracy_score
# 1)
from sklearn.tree import DecisionTreeClassifier

# DecisionTreeClassifier version 1 model
dct1_model = DecisionTreeClassifier()

dct1_model.fit(x_train, y_train)
dct1_pred = dct1_model.predict(x_test)
dct1_pred_score_1 = accuracy_score(y_test, dct1_pred)

print("Prediction list of first dct: ", dct1_pred[:5])
print("Score of first dct: ", dct1_pred_score_1)


# 2)
from sklearn.ensemble import RandomForestClassifier

# RandomForestClassifier vesion 1 model
rfc_1_model = RandomForestClassifier()

rfc_1_model.fit(x_train, y_train)
rfc_1_pred = rfc_1_model.predict(x_test)

rfc_1_pred_score_1 = accuracy_score(y_test, rfc_1_pred)

print("Prediction list of rfc: ", rfc_1_pred[:5])
print("Score of rfc: ", rfc_1_pred_score_1)


# 3)
from sklearn.ensemble import StackingClassifier

estimators = [('dt', DecisionTreeClassifier()),
              ('rf', RandomForestClassifier())]

stc1_model = StackingClassifier(estimators=estimators)

stc1_model.fit(x_train, y_train)
stc1_pred = stc1_model.predict(x_test)

stc_1_pred_score_1 = accuracy_score(y_test, stc1_pred)

print("Prediction list of Stack classifier model: ", stc1_pred[:5])
print("Score of Stack classifier model: ", stc_1_pred_score_1)
```

Summarizing your analysis and observation

```python
'''First model is simple Decision Tree Classifier. It is having not bad accuracy score: 0.82. How we can see Decision Tree Classifier is really good at this dataset. SECOND MODEL IS  RandomForestClassifier vesion 1 model we can see that Random Forest Classifier is better than Decision Tree here, because it is having 0.85 of accuracy score.'''
```

```python
#DATA PREPROCESSING
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
import random
from time import perf_counter
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_theme()
plt.rcParams["figure.figsize"] = (20, 10)
#def seed():
#    return random.getrandbits(32)
def seed():
    return 3198926009
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
file_path = '/content/gdrive/My Drive/DATA/Video_games_esrb_rating.csv'
testfile_path='/content/gdrive/My Drive/DATA/test_esrb.csv'
# Read the CSV file into a Pandas DataFrame
train_pdf = pd.read_csv(file_path)
test_pdf = pd.read_csv(testfile_path)

#Feature engineering
#for checking missing data
train_pdf.isna().any().to_numpy().nonzero()

train_pdf= train_pdf.query("no_descriptors == 0")
train_pdf.pop(train_pdf.no_descriptors.name)
test_pdf= test_pdf.query("no_descriptors == 0")
test_pdf.pop(test_pdf.no_descriptors.name)

train_pdf.esrb_rating.sort_index().value_counts()
train_pdf.esrb_rating.value_counts().sort_index().plot(kind="bar");

#We convert the rating to the minimum age recommended by ESRB. This will allow us to reason about and present the data a bit better.
rating_to_age = {"E": 0, "ET": 10, "T": 13, "M": 17}
train_pdf.esrb_rating = train_pdf.esrb_rating.transform(lambda x: rating_to_age[x])
test_pdf.esrb_rating = test_pdf.esrb_rating.transform(lambda x: rating_to_age[x])

#Let's have a look at the data's distribution

train_pdf.esrb_rating.sort_index().value_counts()
train_pdf.esrb_rating.value_counts().sort_index().plot(kind="bar");


# split the data
x_train = np.array(train_pdf[train_pdf.columns[2:-1]])
y_train = np.array(train_pdf[train_pdf.columns[-1]])

x_test = np.array(test_pdf[test_pdf.columns[2:-1]])
y_test = np.array(test_pdf[test_pdf.columns[-1]])


#Model selection
#First model: linear regression 1
train_data = train_pdf.copy().sample(frac=1, random_state=seed())
train_data = train_data.drop(train_data.title.name, axis=1)
test_pdf.pop(test_pdf.title.name)
train_label = train_data.pop(train_data.esrb_rating.name)
test_label = test_pdf.pop(test_pdf.esrb_rating.name)

#model training
all_label = pd.concat([test_label, train_label])
ntrain_label = (train_label - all_label.mean()) / all_label.std()

start = perf_counter()
model = LinearRegression().fit(train_data, ntrain_label)
finish = perf_counter() - start

pred = model.predict(test_pdf)
pred = (pred * all_label.std()) + all_label.mean()

ages = np.array(list(rating_to_age.values()), dtype=np.float32)

#model evaluation
def round_rating(r):
    return int(ages[np.argmin(np.abs(r - ages))])

rr = np.vectorize(round_rating)
pred = rr(pred)

print(f"Training time: {finish}")
print(classification_report(test_label, pred))

# 2) CLASSIFICATION
from sklearn.ensemble import RandomForestClassifier

# RandomForestClassifier vesion 1 model
rfc_1_model = RandomForestClassifier()

rfc_1_model.fit(x_train, y_train)
rfc_1_pred = rfc_1_model.predict(x_test)

rfc_1_pred_score_1 = accuracy_score(y_test, rfc_1_pred)
print("result of RandomForestClassifier vesion 1 model")
print("Prediction list of rfc: ", rfc_1_pred[:5])
print("Score of rfc: ", rfc_1_pred_score_1)



```

```python
#Summarizing model's performance by evaluation metrices
rf = RandomForestClassifier(random_state=seed()).fit(train_data, train_label)
best_model = rf
sns.heatmap(
    confusion_matrix(test_label, best_model.predict(test_pdf)),
    annot=True,
    xticklabels=rating_to_age.values(),
    yticklabels=list(reversed(list(rating_to_age.values()))),
    fmt="d",
);


```
